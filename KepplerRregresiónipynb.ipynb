```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
semiejemayor = np.array([1, 0.387, 0.723, 1.000, 1.524, 5.203, 9.539, 19.182, 30.058, 39.439])
periodo = np.array([1 ,0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 247.7])

semiejemayor_cubo = semiejemayor**3
periodo_cuadrado = periodo**2

plt.plot(semiejemayor, periodo)

lr = stats.linregress(semiejemayor_cubo, periodo_cuadrado)
b = lr.intercept
m = lr.slope
Pearson_cuadrado = lr.rvalue**2

plt.plot(b + m * semiejemayor_cubo, periodo_cuadrado )
plt.plot(semiejemayor_cubo, periodo_cuadrado, 'o')
plt.title('Periodo al cuadrado vs Semieje Mayor al cubo')
plt.xlabel('Semieje Mayor (a^3)')
plt.ylabel('Periodo (T^2)')
plt.grid()
plt.show()


print('La constante de proporcionalidad K, de acuerdo a la tercera Ley de Keppley tuvo un valor experimental de:', m)
print('El intercepto b, que deberia ser aproximadamente 0 (en un modelo ideal). El valor experimental obteniedo fue:', b)
print('El coeficiente de correlación de Pearson fue:', Pearson_cuadrado, 'Lo que es muy cercado a 1')
print('Demostrando el comportamiento lineal del fenomeno')
```
